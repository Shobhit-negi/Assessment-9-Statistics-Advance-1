#Q13- What is the Central Limit Theorem (CLT), and why is it critical in statistics?

This is crucial because it allows statisticians to apply statistical methods designed for normal distributions to a wide range of problems, even when the underlying data isn't normally distributed



#Q14- How does the Central Limit Theorem relate to the normal distribution?

#Q15- What is the application of Z statistics in hypothesis testing?

The z-statistic helps assess the likelihood of observing the sample data if the null hypothesis is true.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1)
    p (float): The probability of success (x = 1), 0 <= p <= 1

    Returns:
    float: Probability of x
    """
    if x not in [0, 1]:
        return 0.0  # Invalid input
    return p if x == 1 else (1 - p)

# Example usage
print("P(X=0):", bernoulli_pdf(0, 0.7))
print("P(X=1):", bernoulli_pdf(1, 0.7))
```

```python
#9 import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a given dataset.

    Parameters:
        data (list or numpy array): The input dataset

    Returns:
        numpy array: Z-scores of the dataset
    """
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores
```

```python
#19- <
5 Implement hypothesis testing using Z-statistics for a sample dataset5
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Args:
        sample_data (list or np.array): sample observations
        population_mean (float): hypothesized population mean (H0)
        population_std (float): known population std deviation (σ)
        alpha (float): significance level (default=0.05)

    Returns:
        Z-score, p-value, and test conclusion
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-score
    z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    if p_value < alpha:
        conclusion = "Reject H0: Significant difference."
    else:
        conclusion = "Fail to reject H0: No significant difference."

    return z, p_value, conclusion
```

```python
#20- 9"5 Create a confidence interval for a dataset using Python and interpret the result5
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.

    Parameters:
        data (list or np.array): sample data
        confidence (float): confidence level (default = 0.95)

    Returns:
        (lower bound, upper bound, sample mean)
    """
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # standard error of the mean
    margin = stats.t.ppf((1 + confidence) / 2.0, n - 1) * sem
    return mean - margin, mean + margin, mean
```

```python
#21 - 9<5 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean5
import numpy as np
import scipy.stats as stats

def generate_data_and_ci(mean=100, std=15, n=50, confidence=0.95):
    """
    Generate normal data, compute and interpret confidence interval for the mean.
    """
    # Step 1: Generate data
    np.random.seed(1)
    data = np.random.normal(loc=mean, scale=std, size=n)

    # Step 2: Compute sample statistics
    sample_mean = np.mean(data)
    sample_sem = stats.sem(data)  # standard error of the mean

    # Step 3: Compute confidence interval
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * sample_sem
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")
# Run the function
generate_data_and_ci()
```

```python
#23 - 95 Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution5
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def poisson_cdf(lambda_val=3, max_k=15):
    """
    Calculate and plot the CDF of a Poisson distribution with rate λ.
    """
    # Discrete range of k values
    k_values = np.arange(0, max_k + 1)

    # CDF values
    cdf_values = poisson.cdf(k_values, mu=lambda_val)

    # Print example interpretation
    print(f"P(X ≤ 4) when λ = {lambda_val} is: {poisson.cdf(4, mu=lambda_val):.4f}")

    # Plot
    plt.step(k_values, cdf_values, where='post', label=f'Poisson(λ={lambda_val}) CDF', color='blue')
    plt.title(f'Poisson CDF (λ = {lambda_val})')
    plt.xlabel('k (number of events)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.xticks(k_values)
    plt.ylim(0, 1.05)
    plt.legend()
    plt.show()

# Example usage
poisson_cdf(lambda_val=3)
```

```python
#24- 95 Simulate a random variable using a continuous uniform distribution and calculate its expected value5
import numpy as np

def simulate_uniform(a=10, b=20, n=1000):
    """
    Simulate n samples from a continuous uniform distribution U(a, b)
    and calculate the expected value.
    """
    # Step 1: Simulate data
    samples = np.random.uniform(low=a, high=b, size=n)

    # Step 2: Calculate theoretical and empirical expected values
    theoretical_mean = (a + b) / 2
    empirical_mean = np.mean(samples)

    print(f"Theoretical Expected Value: {theoretical_mean:.2f}")
    print(f"Empirical Mean from Simulation: {empirical_mean:.2f}")

# Example: simulate from U(10, 20)
simulate_uniform(a=10, b=20)
```

```python
#25- 9B5 Write a Python program to compare the standard deviations of two datasets and visualize the difference5
import numpy as np
import matplotlib.pyplot as plt

def compare_stddevs():
    # Step 1: Generate two datasets
    np.random.seed(1)
    data1 = np.random.normal(loc=50, scale=5, size=1000)   # std ~ 5
    data2 = np.random.normal(loc=50, scale=10, size=1000)  # std ~ 10

    # Step 2: Calculate standard deviations
    std1 = np.std(data1, ddof=1)
    std2 = np.std(data2, ddof=1)

    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")

    # Step 3: Visualize histograms
    plt.figure(figsize=(12, 5))

    # Histograms
    plt.subplot(1, 2, 1)
    plt.hist(data1, bins=30, alpha=0.6, label='Dataset 1 (σ ~ 5)', color='skyblue')
    plt.hist(data2, bins=30, alpha=0.6, label='Dataset 2 (σ ~ 10)', color='orange')
    plt.title('Histogram Comparison')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)

    # Bar chart of standard deviations
    plt.subplot(1, 2, 2)
    plt.bar(['Dataset 1', 'Dataset 2'], [std1, std2], color=['skyblue', 'orange'])
    plt.title('Standard Deviation Comparison')
    plt.ylabel('Standard Deviation')
    plt.grid(True, axis='y', linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# Run the function
compare_stddevs()
```

```python
#26- 9*5 import numpy as np
import matplotlib.pyplot as plt

def range_and_iqr(mean=100, std=15, size=1000):
    # Step 1: Generate normal data
    np.random.seed(5)
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Calculate Range
    data_range = np.max(data) - np.min(data)

    # Step 3: Calculate IQR (Q3 - Q1)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1

    print(f"Range: {data_range:.2f}")
    print(f"IQR (Interquartile Range): {iqr:.2f}")

    # Step 4: Visualize with boxplot
    plt.boxplot(data, vert=False)
    plt.title("Boxplot of Normally Distributed Data")
    plt.xlabel("Value")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Run the function
range_and_iqr()
```

```python
#27- 9E5 Implement Z-score normalization on a dataset and visualize its transformation5
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization():
    # Step 1: Generate synthetic data (normal distribution)
    np.random.seed(5)
    data = np.random.normal(loc=100, scale=20, size=1000)

    # Step 2: Apply Z-score normalization
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    # Step 3: Print summary
    print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
    print(f"Normalized Data -> Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")

    # Step 4: Visualize
    plt.figure(figsize=(12, 5))

    # Histogram: original data
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, color='skyblue', edgecolor='black')
    plt.title("Original Data Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

    # Histogram: normalized data
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=30, color='orange', edgecolor='black')
    plt.title("Z-score Normalized Data")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()

# Run the function
z_score_normalization()
```

```python
#28- Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution


import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution(mean=0, std=1, size=1000):
    # Step 1: Generate normal data
    np.random.seed(42)
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # default: Fisher's definition (normal = 0)

    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis: {data_kurtosis:.4f}")

    # Optional interpretation
    if abs(data_skewness) < 0.5:
        print("Interpretation: Data is approximately symmetric.")
    else:
        print("Interpretation: Data is skewed.")

    if abs(data_kurtosis) < 1:
        print("Interpretation: Kurtosis is close to normal (mesokurtic).")
    elif data_kurtosis > 1:
        print("Interpretation: Leptokurtic (heavy tails).")
    else:
        print("Interpretation: Platykurtic (light tails).")

# Run the function
analyze_normal_distribution()
```
